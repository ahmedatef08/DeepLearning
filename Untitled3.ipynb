```python
import os
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, roc_curve, auc
```

```python
# 1. Data Preprocessing
DATA_DIR = r"C:\Users\Ahmed Atef\OneDrive\Desktop\Pr-Deep\archive1\lfw-deepfunneled\LWF"  # Replace with your dataset path
BATCH_SIZE = 32
IMG_SIZE = (224, 224)

train_datagen = ImageDataGenerator(
    rescale=1.0 / 255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training"
)

val_generator = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation"
)
```

```python
# 2. Define ResNet Architecture from Scratch
def conv_block(x, filters, kernel_size, strides=1):
    x_skip = x
    x = layers.Conv2D(filters, kernel_size, strides=strides, padding="same", kernel_initializer="he_normal")(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Conv2D(filters, kernel_size, strides=1, padding="same", kernel_initializer="he_normal")(x)
    x = layers.BatchNormalization()(x)

    if strides != 1 or x_skip.shape[-1] != filters:
        x_skip = layers.Conv2D(filters, 1, strides=strides, kernel_initializer="he_normal")(x_skip)
        x_skip = layers.BatchNormalization()(x_skip)

    x = layers.Add()([x, x_skip])
    x = layers.ReLU()(x)
    return x

def build_resnet(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)
    x = layers.Conv2D(64, 7, strides=2, padding="same", kernel_initializer="he_normal")(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D(3, strides=2, padding="same")(x)

    # ResNet Blocks
    x = conv_block(x, 64, 3)
    x = conv_block(x, 64, 3)

    x = conv_block(x, 128, 3, strides=2)
    x = conv_block(x, 128, 3)

    x = conv_block(x, 256, 3, strides=2)
    x = conv_block(x, 256, 3)

    x = conv_block(x, 512, 3, strides=2)
    x = conv_block(x, 512, 3)

    x = layers.GlobalAveragePooling2D()(x)
    outputs = layers.Dense(num_classes, activation="softmax")(x)

    model = models.Model(inputs, outputs)
    return model

# Initialize ResNet
num_classes = train_generator.num_classes
input_shape = IMG_SIZE + (3,)
model = build_resnet(input_shape, num_classes)
model.compile(optimizer=optimizers.Adam(learning_rate=0.001),
              loss="categorical_crossentropy",
              metrics=["accuracy"])
```

```python
# 3. Training the Model
EPOCHS = 5
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS
)
# 4. Save the Model
model.save("resnet_from_scratch.h5")
```

```python
# 5. Display Final Accuracy
train_acc = history.history['accuracy'][-1] * 100
val_acc = history.history['val_accuracy'][-1] * 100

print(f"Final Accuracy: {train_acc:.2f}%")
```

```python
# Reset the validation generator
val_generator.reset()

# Predict the classes for the validation set
predictions = model.predict(val_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = val_generator.classes

# Get the class labels corresponding to the true classes in the validation set
class_labels = list(val_generator.class_indices.keys())
actual_labels = [class_labels[i] for i in np.unique(true_classes)]

# Generate the confusion matrix
conf_matrix = confusion_matrix(true_classes, predicted_classes)

# Plot the confusion matrix with increased size
plt.figure(figsize=(15, 12))  # Adjust the size as needed
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=actual_labels)
disp.plot(cmap="Blues", ax=plt.gca())  # Use the current axis to apply the larger size
plt.title("Confusion Matrix", fontsize=18)  # Add a larger title for clarity
plt.xticks(fontsize=12, rotation=45, ha='right')  # Adjust x-axis label size and rotation
plt.yticks(fontsize=12)  # Adjust y-axis label size
plt.show()
```

```python
# Classification Report
print("Classification Report:")
print(classification_report(true_classes, predicted_classes, labels=range(num_classes), target_names=class_labels))
```

```python
# ROC and AUC Visualization with larger figure size
plt.figure(figsize=(15, 12))  # Increased figure size
for i in range(num_classes):
    # Generate true binary labels for class `i`
    true_class_binary = (true_classes == i).astype(int)
    # Get predicted probabilities for class `i`
    predicted_probs = predictions[:, i]
    # Compute FPR and TPR for the class
    fpr, tpr, _ = roc_curve(true_class_binary, predicted_probs)
    # Compute AUC for the class
    roc_auc = auc(fpr, tpr)
    # Plot the ROC curve for the class
    plt.plot(fpr, tpr, label=f"{class_labels[i]} (AUC = {roc_auc:.2f})")

# Plot the diagonal line for random guessing
plt.plot([0, 1], [0, 1], 'k--', lw=2)

# Customize plot labels and title
plt.xlabel("False Positive Rate", fontsize=14)
plt.ylabel("True Positive Rate", fontsize=14)
plt.title("ROC Curves and AUC", fontsize=18)

# Display the legend
plt.legend(loc="lower right", fontsize=12)

# Enable grid for better readability
plt.grid(alpha=0.4)

# Show the plot
plt.show()
```
